### Group by and Aggregating

```python
df_base = pd.read_csv('demo files/zpopulation.csv')
df_base
```

```python
df = df_base[['Country', 'Continent','2022 Population']].copy()
```

```python
df
```

```python
df.columns
```

```python
df.rename(columns={'2022 Population':'Population'},inplace=True)
df
```

```python
df.groupby('Continent').sum()
```
