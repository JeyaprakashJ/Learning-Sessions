Let's say that we want to write an algorithm to find the maximum value in an array. Here are two possible implementations of this algorithm:



```python
def inefficient_max_function(array):
  max_value = array[0]
  for i in range(1, len(array)):
    if array[i] > max_value:
      max_value = array[i]
  return max_value
```

```python
def efficient_max_function(array):
  max_value = array[0]
  for i in range(1, len(array)):
    max_value = max(max_value, array[i])
  return max_value
```

The inefficient_max_function algorithm has a linear time complexity, while the efficient_max_function algorithm has a constant time complexity. This is because the inefficient_max_function algorithm compares each element of the array to the current maximum value, while the efficient_max_function algorithm only compares the current element to the maximum value.
